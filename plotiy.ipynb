```python
def plot_candlestick(df):
    candleStick = go.Candlestick(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close'], name='Candlesticks')
    fig = go.Figure(data=[candleStick])
    fig.update(layout_xaxis_rangeslider_visible=False)
    return fig
```

```python
fig = plot_candlestick(fdr.DataReader('AAPL', start='2023'))
fig.show()
```

```python
def plot_MDD(df):
    df_dd, list_info = fn.mdd_fn(df)
    dd= go.Scatter(x=df_dd.index, y=df_dd['drawdown'], name='MDD', fill='tozeroy')
    fig = go.Figure(dd)
    return fig
```

```python
def plot_MACD(df):
    df_macd = fn.macdOscillator_fn(df, short_N=9, long_N=26, signal_N=13)
    macd = go.Scatter(x=df_macd.index, y=df_macd['MACD'], name='MACD')
    signal = go.Scatter(x=df_macd.index, y=df_macd['Signal'], name='Signal')
    ocillator = go.Bar(x=df_macd.index, y=df_macd['MACD Oscillator'], name='Oscillator')
    
    fig = go.Figure(data=[macd, signal, ocillator])
    return fig

plot_MACD(fdr.DataReader('AAPL', start='2023'))
```

```python
def plot_RSI(df):
    df_rsi = fn.rsi_fn(df, rsi_period=14)
    rsi = go.Scatter(x=df_rsi.index, y=df_rsi['RSI'], name='RSI')
    fig = go.Figure(rsi)
    
    rsi_overbought = go.Scatter(x=df_rsi.index, y=[70]*len(df_rsi), name='Overbought', line=dict(color='red', dash='dash'))
    rsi_oversold = go.Scatter(x=df_rsi.index, y=[30]*len(df_rsi), name='Oversold', line=dict(color='blue', dash='dash'))
    fig.add_trace(rsi_overbought)
    fig.add_trace(rsi_oversold)
        
    return fig

plot_RSI(fdr.DataReader('AAPL', start='2023'))
```

```python
def plot_volume(df):
    volume = go.Bar(x=df.index, y=df['Volume'], name='Volume')
    fig = go.Figure(volume)
    return fig

plot_volume(fdr.DataReader('AAPL', start='2023'))
```
