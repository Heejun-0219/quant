```python
import chart_fn as cf
from plotly import subplots
import plotly.graph_objects as go
```

```python
def tradingRoom_fn(market, stockName, startDate = None, endDate = None, short_N = 9, long_N = 26, signal_N = 13, rsi_N = 14):
    print("Trading Room")

    cf.todayMarketData()
    
    df = cf.stockDataReader_fn(market=market, stockName=stockName, startDate=startDate, endDate=endDate)

    df_dd, list_info = cf.mdd_fn(df)
    df_macd = cf.macdOscillator_fn(df, short_N=short_N, long_N=long_N, signal_N=signal_N)
    df_rsi = cf.rsi_fn(df, rsi_period=rsi_N)
    
    candleStick = go.Candlestick(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close'], name='Candle Stick')
    mdd = go.Scatter(x=df_dd.index, y=df_dd['drawdown']*100, name='Draw down', fill='tozeroy')
    macd = go.Scatter(x=df_macd.index, y=df_macd['MACD'], mode='lines', name='MACD')
    signal = go.Scatter(x=df_macd.index, y=df_macd['Signal'], mode='lines', name='Signal')
    rsi = go.Scatter(x=df_rsi.index, y=df_rsi['RSI'], mode='lines', name='RSI')
    rsi_high = go.Scatter(x=df_rsi.index, y=[70]*len(df_rsi.index), mode='lines', name='Overbought', line=dict(color='blue', width=1, dash='dash'), showlegend=False)
    rsi_low = go.Scatter(x=df_rsi.index, y=[30]*len(df_rsi.index), mode='lines', name='Oversold', line=dict(color='red', width=1, dash='dash'), showlegend=False)
    vol = go.Bar(x=df.index, y=df['Volume'], name='Volume', marker=dict(color='blue'))
    
    fig = subplots.make_subplots(rows=5, cols=1, vertical_spacing=0.01, shared_xaxes=True, 
                                 subplot_titles=('<b>{}</b>'.format(stockName),'Draw down', 'MACD Oscillator', 'RSI', 'Volume'))
    fig.append_trace(candleStick, row=1, col=1)
    fig.append_trace(mdd, row=2, col=1)
    fig.append_trace(macd, row=3, col=1)
    fig.append_trace(signal, row=3, col=1)
    fig.append_trace(rsi, row=4, col=1)
    fig.add_trace(rsi_high, row=4, col=1)
    fig.add_trace(rsi_low, row=4, col=1)
    fig.append_trace(vol, row=5, col=1)
    
    fig.update_xaxes(rangeslider_visible=False)
    fig.update_layout(width=1000, height=800)
    fig.update_layout(legend_title_text='Contents', title='{}, Max : {}, Min : {}, MDD : {:.2f}% Period : {} ~ {}'.format(stockName, list_info[0], list_info[1], list_info[2]*100, list_info[3], list_info[4]))
    
    fig.show()

tradingRoom_fn(market='nyse', stockName='CPNG')
```

```python
tradingRoom_fn(market='kosdaq', stockName='마녀공장')
```
