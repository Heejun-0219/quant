```python
def get_ema(df, N):
    weight = 2 / (N + 1)
    df['EWM'] = df['Close'][0]
    for index in range(1, len(df)):
        df['EWM'][index] = df['Close'][index] * weight + df['EWM'][index - 1] * (1 - weight)
    return df
```

```python
df = fdr.DataReader('005930', '2021-01-01', '2023-12-31')[['Close', 'Volume']]
df['short_EWM'] = df['Close'].ewm(span=9, adjust=False).mean()
df['long_EWM'] = df['Close'].ewm(span=26, adjust=False).mean()

df['MACD'] = df['short_EWM'] - df['long_EWM']
df['Signal'] = df['MACD'].ewm(span=13, adjust=False).mean()
df['MACD Oscillator'] = df['MACD'] - df['Signal']
df
```

```python
def get_macd(df, short_N=9, long_N=26, signal_N=13):
    df['short_EWM'] = df['Close'].ewm(span=short_N, adjust=False).mean()
    df['long_EWM'] = df['Close'].ewm(span=long_N, adjust=False).mean()
    df['MACD'] = df['short_EWM'] - df['long_EWM']
    df['Signal'] = df['MACD'].ewm(span=signal_N, adjust=False).mean()
    df['MACD Oscillator'] = df['MACD'] - df['Signal']
    return df
```

```python
df = fdr.DataReader('005930', '2021-01-01', '2023-12-31')[['Close', 'Volume']]
df = get_macd(df)
df
```

```python
def macdOscillator_fn(df, short_N=9, long_N=26, signal_N=13):
    df = df[['Close']].copy() 
    
    df['Short'] = df['Close'].ewm(span=short_N, adjust=False).mean()
    df['Long'] = df['Close'].ewm(span=long_N, adjust=False).mean()
    
    df['MACD'] = df['Short'] - df['Long']
    df['Signal'] = df['MACD'].ewm(span=signal_N, adjust=False).mean()
    df['MACD Oscillator'] = df['MACD'] - df['Signal']
    
    return df[['MACD', 'Signal', 'MACD Oscillator']]
```

```python
df = fdr.DataReader('005930', '2021-01-01', '2023-12-31')
df = macdOscillator_fn(df)
```
